```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d

from scipy.optimize import minimize
```

```python
def objective(w): # 3-d view
    return w[0]**2 + w[1]**2 #+ w[2]**2
```

```python
def constraint(w): # along the 2-d plane we have a line
    return w[0] + 2*w[1] - 1
```

```python
fig = plt.figure(figsize=(12,10))
ax = fig.add_subplot(111, projection='3d')
X, Y = np.mgrid[-1:1:30j, -1:1:30j]
Z = X**2 + Y**2
sur = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='jet', lw=0.5)
ax.contour(X, Y, Z, 10, offset=-1, cmap='jet', linestyles = "solid")
ax.contour(X, Y, Z, 10, colors="k", linestyles = "solid")
plt.show()
```

```python
w_start = [3, 5]
bounds  = ((-10, 10), (-10, 10))
solution = minimize(objective, w_start, bounds=bounds)
solution.x
```

```python
w_start = [3, 5]
bounds  = ((-10, 10), (-10, 10))
constraint = [{'type': 'eq', 'fun': constraint}]
solution = minimize(objective, w_start, bounds=bounds, constraints=constraint)
solution.x
```
