```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_func(func):
  x_values = np.linspace(-10, 10, 1000)
  y_values = [func(x) for x in x_values]
  plt.plot(x_values, y_values)
```

```python
plot_func(lambda x:x**2)
```

```python
def plot_derivative(func):
  x_values = np.linspace(-10, 10, 1000)
  delta_x = 0.0001
  y_values = (func(x_values + delta_x) - func(x_values)) / delta_x
  plt.plot(x_values, y_values)
```

```python
plot_derivative(lambda x:x**3)
```

```python
# Gradient Descent

# Most easy version - just to get a bigger picture of it


# Given a function f and a starting point x0, try to find the minima
def gradient_descent(func, x0):
  list_of_point = []
  x = x0
  delta = 0.0001

  for i in range(100):
    derivative = (func(x + delta) - func(x)) / delta
    x = x - 0.1*derivative  # 0.1 is learning rate, without this result will fluctuate too much, oscillation will happen like mad
    list_of_point.append(x)
  return list_of_point
```

```python
gradient_descent(lambda x:x**2, 0.5)
# result will come down to minima as we increase the number of iteration, 0 is point of minima for f(x) = x**2
```

```python
gradient_descent(lambda x:x**2 + 10*x + 3, 100)[-1]
# -5 seems to be the minima
```

```python
plot_func(lambda x:x**2 + 10*x + 3)
```

```python
plot_derivative(lambda x:x**2 + 10*x + 3)
# At the point of minima (-5) we can see that derivative is 0, left of it is negative and right of it positive
```
